```python
# Menghitung kuartil pertama dari nilai-nilai dalam kolom 'math_score' dalam DataFrame df_noid
q1 = df_noid['math_score'].quantile(q=0.25)

# Menghitung kuartil ketiga dari nilai-nilai dalam kolom 'math_score' dalam DataFrame df_noid
q3 = df_noid['math_score'].quantile(q=0.75)

# Menghitung rentang interkuartil dari nilai-nilai dalam kolom 'math_score' dalam DataFrame df_noid
iqr = q3 - q1

# Membuat masker boolean untuk menandai apakah nilai 'math_score' merupakan outlier
iso = (df_noid['math_score'] < q1 - 1.5*iqr) | (df_noid['math_score'] > q3 + 1.5*iqr)

# Membuat masker boolean untuk menandai apakah nilai 'math_score' merupakan extreme outlier
iseo = (df_noid['math_score'] < q1 - 3*iqr) | (df_noid['math_score'] > q3 + 3*iqr)

# Membuat DataFrame baru yang berisi kolom 'gender' dan 'math_score', serta dua kolom baru 'is_outlier' dan
#'is_extreme_outlier' yang menunjukkan apakah nilai 'math_score' merupakan outlier atau extreme outlier
df1 = df_noid[['gender', 'math_score']].assign(is_outlier=iso, is_extreme_outlier=iseo)

# Memilih baris-baris di mana nilai 'math_score' merupakan outlier atau extreme outlier dan menampilkannya
df1.loc[df1['is_outlier'] | df1['is_extreme_outlier']]
```

```python
#Menampilkan diagram pie
education = ('masters degree','bachelors degree','associate degree','some college','some high school','high school')
votes = (76, 73, 70, 68, 64, 62)
colors = ('#772a7d','#d55bde','#c9a3cc','#2f1830','#db0b96','#75053d')
explode = (0,0,0,0,0,0.1)
plt.title('Writing Score Based On Parental Level Of Education')
plt.pie(
        votes,
        labels=education,
        autopct='%1.lf%%',
        colors=colors,
        explode=explode,
        shadow=True
        )
plt.show()
```

```python
#menamplikan diagram batang secara vertikal
gender = ('Female','Male')
populations = (518, 482)
df = pd.DataFrame({
    'Gender' : gender,
    'Populations' : populations})
df.sort_values(by='Populations', inplace=True)

x_coords = np.arange(len(df))
colors = ['#d55bde' for _ in range(len(df))]
colors[-2] = '#2f1830'
plt.figure(figsize=(4,5))
plt.bar(x_coords, df['Populations'], tick_label=df['Gender'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Population')
plt.title('Gender Populations')
plt.show()
```
